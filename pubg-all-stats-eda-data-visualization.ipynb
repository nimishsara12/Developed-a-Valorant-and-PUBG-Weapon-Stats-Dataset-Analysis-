# Weapons are classified into different types.
# The following Plot shows different types of Weapons And Frequency of their availability : From the Plot we can say Assualt Riffles are more Frequently available.

```python
plt.figure(figsize=(15,5))
plt.hist(df1['Weapon Type'],bins=15,color='lightgreen')
plt.xlabel('Weapon Type')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.grid(True)
```

# The following Plot shows the Magazine Capacity of each Weapon Type : From the Plot we can analyze 30 Magazine Guns are more in number. 

```python
plt.figure(figsize=(15,5))
plt.hist(df1['Magazine Capacity'],bins=20,color='b')
plt.xlabel('Magazine Capacity')
plt.ylabel('No. of Guns')
plt.title('Histogram')
plt.grid(True)
```

# Weapon Type Vs It's Range in Meters(i.e., How far the Bullet Travels).

```python
plt.figure(figsize=(15,5))
sns.barplot(x = 'Weapon Type', y = 'Range', data = pd.read_csv(r'/kaggle/input/pubg-weapon-stats/pubg-weapon-stats.csv'))
```

# The following Plot shows the comparision between Range and Damage of different Weapon Type (HeadDamage without Helmet) :
# Long Range and High Damage Guns are more Efficient and are used more Frequently : From this we can visualize Assault Riffle is more Consistent and used more Frequently.

```python
plt.figure(figsize=(15,5))
plt.bar(df1['Weapon Type'], df1['Range'], label='Range', color='g')
plt.bar(df1['Weapon Type'], df1['HDMG_0'], label='Damage', color='r')
plt.xlabel('Weapon Type')
plt.ylabel('Range & Head Damage')
plt.title('Bar Chart')
plt.legend()
plt.grid(True)
```

# First for Weapon Type we Visualize each of the following individually :
# 1) Damage
# 2) Range
# 3) Bullet Speed

# Then we Visualize any Two Categories of Weapon Types which are more Efficient.

# 1) Weapon Type vs Damage

```python
df2=df1.groupby('Weapon Type')['Damage'].mean().reset_index()
df2
```

```python
plt.figure(figsize=(7,7))
sns.barplot(y = 'Weapon Type', x = 'Damage', data = df2)
```

# 2) Weapon Type vs Range

```python
df3=df1.groupby('Weapon Type')['Range'].mean().reset_index()
df3
```

```python
plt.figure(figsize=(7,7))
sns.barplot(y = 'Weapon Type', x = 'Range', data = df3)
```

# 3) Weapon Type vs Bullet Speed

```python
df4=df1.groupby('Weapon Type')['Bullet Speed'].mean().reset_index()
df4
```

```python
plt.figure(figsize=(7,7))
sns.barplot(y = 'Weapon Type', x = 'Bullet Speed', data = df4)
```

# 4) Weapon Type vs Rate of Fire

```python
df5=df1.groupby('Weapon Type')['Rate of Fire'].mean().reset_index()
df5
```

```python
plt.figure(figsize=(7,7))
sns.barplot(y = 'Weapon Type', x = 'Rate of Fire', data = df5)
```
